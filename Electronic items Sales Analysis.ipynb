```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel file
data = pd.read_excel(r"C:\Users\PHANI\OneDrive\Desktop\sample for outlayer.xlsx")

# Display first 5 rows
print(data.head())

# Basic Information
print("\nDataset Information")
print(data.info())

# Total Sales
total_sales = data["Total_Sales"].sum()
print("Total Sales:", total_sales)

# Total Quantity
total_quantity = data["Quantity"].sum()
print("Total Quantity Sold:", total_quantity)

# Sales by Category
category_sales = data.groupby("Category")["Total_Sales"].sum()
print(category_sales)

# Top Selling Products
top_products = data.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False)
print(top_products)

# Bar Chart
plt.figure(figsize=(8,5))
top_products.plot(kind="bar")
plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(6,6))
category_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Category Wise Sales")
plt.ylabel("")
plt.show()

# Highest Selling Product
best_product = top_products.idxmax()
best_sales = top_products.max()

print("Best Selling Product:", best_product)
print("Highest Sales:", best_sales)
```
